```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn import linear_model


```

```python
df = pd.read_csv('canada_per_capita_income.csv')

df.head()
```

```python
new_df = df.copy()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(new_df.year, new_df['per capita income (US$)'], color='blue')
```

```python
model = linear_model.LinearRegression()
model.fit(new_df[['year']], new_df['per capita income (US$)'])
```

```python
model.predict([[2020]])
```

```python
m, c = model.coef_, model.intercept_
```

```python
m,c
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(new_df.year, new_df['per capita income (US$)'], color='red')
plt.plot(new_df.year, m*new_df.year + c, color='blue')
```
